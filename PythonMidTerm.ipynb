```python
##Using AI, I got some help on figuring out how to show the median values of % of students scoring >65 for the 4 selected subjects, comparing the performance in a bar chart side by side for both Bronx Arena HS and Orchard Collegiate##
# prompt: plot bar chart for above cell with median for Percent Scoring 65 or Above, with Bronx Arena in orange and Orchard Collegiate in green and non-overlapping labels for Regents Exam

import matplotlib.pyplot as plt

# Create some data
labels = ['Global History and Geography', 'U.S. History and Government', 'Physical Settings/Earth Science', 'English']
values_bronx = [BrxArena2015HistGeog['Percent Scoring 65 or Above'].median(), BrxArena2015HistGovt['Percent Scoring 65 or Above'].median(), BrxArena2015EarthScience['Percent Scoring 65 or Above'].median(), BrxArena2015English['Percent Scoring 65 or Above'].median()]
values_orchard = [Orchard2015HistGeog['Percent Scoring 65 or Above'].median(), Orchard2015HistGovt['Percent Scoring 65 or Above'].median(), Orchard2015EarthScience['Percent Scoring 65 or Above'].median(), Orchard2015English['Percent Scoring 65 or Above'].median()]

# Define width of each bar
width = 0.25

# Define x-axis positions for each bar
x_bronx = [i - width / 2 for i in range(len(values_bronx))]
x_orchard = [i + width / 2 for i in range(len(values_orchard))]

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create the bars
rect1 = ax.bar(x_bronx, values_bronx, width, color='green', label='Bronx Arena HS')
rect2 = ax.bar(x_orchard, values_orchard, width, color='purple', label='Orchard Collegiate')

# Add labels and title
ax.set_title('Median Percent Scoring 65 or Above on Regents Exams in 2015', fontsize=16)
ax.set_ylabel('% Scoring 65 or Above', fontsize=14)
ax.set_xlabel('Regents Exam Subject', fontsize=14)
ax.set_xticks([i + width / 2 for i in range(len(labels))])
ax.set_xticklabels(labels, rotation=45, ha='right')

# Add non-overlapping labels
for rect in rect1:
    height = rect.get_height()
    ax.annotate('{}'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

for rect in rect2:
    height = rect.get_height()
    ax.annotate('{}'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

# Add legend and show plot
ax.legend()
plt.show()
```

The above bar chart shows that in the year 2015, Bronx Arena High School outperformed Orchard Collegiate Academy in median percentage of students with scores 65 or higher for 3 out of the 4 Regents Exam subjects in which both schools participated.
